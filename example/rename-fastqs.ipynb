```python
from xml.dom import minidom
import pandas as pd
import qiime2
import os
```

```python
lookup = qiime2.Metadata.load('sample-metadata.tsv').get_column('SRA').to_dataframe()
lookup.reset_index(inplace=True)
lookup.set_index('SRA', inplace=True)
lookup = lookup['sample-id']
```

```python
run_to_sample = {}
xml = minidom.parse('SraExperimentPackage.xml')

for run in xml.getElementsByTagName('RUN'):
    sample = run.getElementsByTagName('Member')[0].getAttribute('accession')
    run_to_sample[run.getAttribute('accession')] = lookup[sample]
        
```

```python
orig = [(f.getAttribute('filename'), f.getAttribute('date')) for f in xml.getElementsByTagName("SRAFile") if f.getAttribute('filename').startswith('BCN0')]
```

```python
sorted(list(zip(orig, [int(s[0].split('_')[3][1:]) for s in orig])), key=lambda x: x[1])
```

```python
bc = {}
counter = 0
for file in os.listdir('fastq'):
    name, *_ = file.split('.')
    ident, _, R = name.split('_')
    if ident not in bc:
        counter += 1
        bc[ident] = counter
    
    new = '%s_S%s_L001_R%s_001.fastq.gz' % (run_to_sample[ident], bc[ident], R)
    print(new)
    
    
```
